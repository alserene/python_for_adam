```python
# Get code packages
import tkinter as tk
from tkinter import ttk
from tkinter import *

# Create the window, name it and give it some padding.
window = tk.Tk()
window.title("Runnin.")
window['padx'] = 5
window['pady'] = 5

# Define empty lists to fill with run data.
days = []
distances = []
times = []

# Define functions to run with button presses.
def submit():
    days.append(day.get())
    distances.append(distance.get())
    times.append(time.get())
    print(f"days: {days}\ndistances: {distances}\ntimes: {times}\n")
    
    compute_avg_dist()
    compute_pace()
    
def compute_avg_dist():
    result = sum(distances)/len(distances)
    avg_dist_label.config(text=f"Average Distance: {result:.2f} km")

def compute_pace():
    total_time = sum(times)
    total_distance = sum(distances)
    if total_distance == 0:
        result = 0
    else:
        result = total_time / total_distance
    avg_pace_label.config(text=f"Average Pace: {result:.2f} min/km")

# Input variables.
day = tk.StringVar()
distance = tk.DoubleVar()
time = tk.IntVar()

# Output variables.
avg_dist = tk.DoubleVar()
avg_pace = tk.DoubleVar()

# The input frame - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
input_frame = ttk.LabelFrame(window, relief=tk.RIDGE)
input_frame.grid(row=1, column=1, sticky=tk.E + tk.W + tk.N + tk.S)

# Day label.
day_input_label = ttk.Label(input_frame, text="Day")
day_input_label.grid(row=1, column=1, sticky=tk.W + tk.N)

# Distance label.
dist_input_label = ttk.Label(input_frame, text="Distance")
dist_input_label.grid(row=2, column=1, sticky=tk.W + tk.N)

# Time label.
time_input_label = ttk.Label(input_frame, text="Time")
time_input_label.grid(row=3, column=1, sticky=tk.W + tk.N)

# Day input.
day_input = ttk.Combobox(input_frame, height=4, textvariable=day)
day_input.grid(row=1, column=2, sticky=tk.W)
day_input['values'] = ("Monday", "Tuesday", "Wednesday", "Thurdsay", "Friday", "Saturday", "Sunday")

# Distance input.
dist_input = ttk.Entry(input_frame, textvariable=distance,width=20)
dist_input.grid(row=2, column=2, sticky=tk.W, pady=3)

# Time input.
time_input = tk.Scale(input_frame, variable=time, from_=0, to=60, orient=tk.HORIZONTAL, width=8, length=200)
time_input.grid(row=3, column=2, sticky=tk.W)

# Submit button.
submit_input = ttk.Button(input_frame, text="Submit run", command=submit)
submit_input.grid(row=4, column=2)

# The output frame - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
output_frame = ttk.LabelFrame(window, relief=tk.RIDGE)
output_frame.grid(row=2, column=1, sticky=tk.E + tk.W + tk.N + tk.S)

# Avg distance output.
avg_dist_label = ttk.Label(output_frame, text="Avg. Distance: ")
avg_dist_label.grid(row=1, column=1, sticky=tk.W)

# Pace output.
avg_pace_label = ttk.Label(output_frame, text="Avg. Pace: ")
avg_pace_label.grid(row=2, column=1, sticky=tk.W)

# Keep the GUI app running and responsive to events - - - - - - - - - - - - - - -
window.mainloop()
```

<p>The buttons, text input, drop-down list and slider are all examples of <strong>widgets</strong>. Each type of widget has a different purpose and offers a way for a user (us) to interact with an app.<br><br>Here is some code I found online that gives a nice visual demonstration of all the different widgets that we can use in our Tkinter app.</p>

```python
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog


class Counter_program():
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("tk Examples")
        self.create_widgets()

        self.radio_variable = tk.StringVar()
        self.combobox_value = tk.StringVar()

    def create_widgets(self):
        # Create some room around all the internal frames
        self.window['padx'] = 5
        self.window['pady'] = 5

        # - - - - - - - - - - - - - - - - - - - - -
        # The Buttons frame
        button_frame = ttk.LabelFrame(self.window, text="Buttons", relief=tk.RIDGE)
        button_frame.grid(row=1, column=1, sticky=tk.E + tk.W + tk.N + tk.S)
        
        # Buttons
        tk_button = tk.Button(button_frame, text="do something")
        tk_button.grid(row=1, column=2, sticky=tk.W)
        
        ttk_button = ttk.Button(button_frame, text="do something")
        ttk_button.grid(row=2, column=2, sticky=tk.W)
        
        # Labels
        tk_button_label = ttk.Label(button_frame, text="tk.Button")
        tk_button_label.grid(row=1, column=1, sticky=tk.W, pady=3)
        ttk_button_label = ttk.Label(button_frame, text="ttk.Button")
        ttk_button_label.grid(row=2, column=1, sticky=tk.W, pady=3)
        
        # - - - - - - - - - - - - - - - - - - - - -
        # The Data Input frame
        input_frame = ttk.LabelFrame(self.window, text="Data Input", relief=tk.RIDGE)
        input_frame.grid(row=2, column=1, sticky=tk.E + tk.W + tk.N + tk.S)

        # Text inputs
        my_entry = ttk.Entry(input_frame, width=20)
        my_entry.grid(row=1, column=2, sticky=tk.W, pady=3)
        my_entry.insert(tk.END, "Add some text here ...")

        my_text = tk.Text(input_frame, height=5, width=30)
        my_text.grid(row=2, column=2, sticky=tk.W)
        my_text.insert(tk.END, "Add some text here ...\nand here ...")

        # Spinbox
        my_spinbox = tk.Spinbox(input_frame, from_=0, to=10, width=5, justify=tk.RIGHT)
        my_spinbox.grid(row=3, column=2, sticky=tk.W, pady=3)

        # Scales
        my_scale = tk.Scale(input_frame, from_=0, to=100, orient=tk.HORIZONTAL, width=8, length=200)
        my_scale.grid(row=4, column=2, sticky=tk.W)

        my_scale = ttk.Scale(input_frame, from_=0, to=100, orient=tk.HORIZONTAL, length=200)
        my_scale.grid(row=5, column=2, sticky=tk.W)
        
        # Labels
        entry_label = ttk.Label(input_frame, text="ttk.Entry")
        entry_label.grid(row=1, column=1, sticky=tk.W + tk.N)
        text_label = ttk.Label(input_frame, text="tk.Text")
        text_label.grid(row=2, column=1, sticky=tk.W + tk.N)
        spin_label = ttk.Label(input_frame, text="tk.Spinbox")
        spin_label.grid(row=3, column=1, sticky=tk.W + tk.N)
        scale_label = ttk.Label(input_frame, text="tk.Scale")
        scale_label.grid(row=4, column=1, sticky=tk.W)
        scale_label2 = ttk.Label(input_frame, text="ttk.Scale")
        scale_label2.grid(row=5, column=1, sticky=tk.W)

        # - - - - - - - - - - - - - - - - - - - - -
        # The Lists frame
        list_frame = ttk.LabelFrame(self.window, text="Choose from a list", relief=tk.RIDGE)
        list_frame.grid(row=1, column=2, sticky=tk.E + tk.W + tk.N + tk.S, padx=6)

        # Labels
        listbox_label = ttk.Label(list_frame, text="tk.Listbox")
        listbox_label.grid(row=1, column=1, sticky=tk.W + tk.N)
        combobox_label = ttk.Label(list_frame, text="ttk.Combobox")
        combobox_label.grid(row=2, column=1, sticky=tk.W + tk.N)

        # Listbox
        my_listbox = tk.Listbox(list_frame, height=4)
        for item in ["one", "two", "three", "four"]:
            my_listbox.insert(tk.END, "Choice " + item)
        my_listbox.grid(row=1, column=2)

        # Combobox
        self.combobox_value = tk.StringVar()
        my_combobox = ttk.Combobox(list_frame, height=4, textvariable=self.combobox_value)
        my_combobox.grid(row=2, column=2)
        my_combobox['values'] = ("Choice one", "Choice two", "Choice three", "Choice four")
        my_combobox.current(0)
        
    # - - - - - - - - - - - - - - - - - - - - -
        # The Choice frame
        choice_frame = ttk.LabelFrame(self.window, text="Choices", relief=tk.RIDGE, padding=6)
        choice_frame.grid(row=2, column=2, padx=6, sticky=tk.E + tk.W + tk.N + tk.S)

        # Labels
        checkbox_label = ttk.Label(choice_frame, text="ttk.Checkbutton")
        checkbox_label.grid(row=1, rowspan=3, column=1, sticky=tk.W + tk.N)
        entry_label = ttk.Label(choice_frame, text="ttk.Radiobutton")
        entry_label.grid(row=4, rowspan=3, column=1, sticky=tk.W + tk.N)

        # Checkbutton
        checkbutton1 = ttk.Checkbutton(choice_frame, text="On-off switch 1")
        checkbutton1.grid(row=1, column=2, sticky=tk.W)
        checkbutton2 = ttk.Checkbutton(choice_frame, text="On-off switch 2")
        checkbutton2.grid(row=2, column=2, sticky=tk.W)
        checkbutton3 = ttk.Checkbutton(choice_frame, text="On-off switch 3")
        checkbutton3.grid(row=3, column=2, sticky=tk.W)

        # Radiobutton
        self.radio_variable = tk.StringVar()
        self.radio_variable.set("0")
        radiobutton1 = ttk.Radiobutton(choice_frame, text="Choice One of three", variable=self.radio_variable, value="0")
        radiobutton2 = ttk.Radiobutton(choice_frame, text="Choice Two of three", variable=self.radio_variable, value="1")
        radiobutton3 = ttk.Radiobutton(choice_frame, text="Choice Three of three", variable=self.radio_variable, value="2")
        radiobutton1.grid(row=4, column=2, sticky=tk.W)
        radiobutton2.grid(row=5, column=2, sticky=tk.W)
        radiobutton3.grid(row=6, column=2, sticky=tk.W)

        # - - - - - - - - - - - - - - - - - - - - -
        # Menus
        menubar = tk.Menu(self.window)

        filemenu = tk.Menu(menubar, tearoff=0)
        filemenu.add_command(label="Open", command=filedialog.askopenfilename)
        filemenu.add_command(label="Save", command=filedialog.asksaveasfilename)
        filemenu.add_separator()
        filemenu.add_command(label="Exit", command=self.window.quit)
        menubar.add_cascade(label="File", menu=filemenu)

        self.window.config(menu=menubar)

        # - - - - - - - - - - - - - - - - - - - - -
        # Quit button in the lower right corner
        quit_button = ttk.Button(self.window, text="Quit", command=self.window.destroy)
        quit_button.grid(row=3, column=2, sticky=tk.E)

# Create the entire GUI program
program = Counter_program()

# Start the GUI event loop
program.window.mainloop()
```
